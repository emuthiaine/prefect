```python
import prefect
import pandas as pd
```

```python
from prefect import task, flow, deployments
```

### define data processing function

```python
@task()
def read_data():
    df = pd.read_csv('route_fare_stops.csv')
    return df
```

```python
@task()
def calculate_mean(df):
    df_summary = df.groupby('route_name')['fare'].mean().reset_index()
    return df_summary
```

```python
@task()
def save_to_csv(df_summary):
    df_summary.to_csv('summary.csv')
```

```python
@flow()
def data_processing_flow():
    data = read_data()
    summary = calculate_mean(data)
    save_to_csv(summary)
```

```python
if __name__ == "__main__":
    data_processing_flow.run()
```

```python
save_to_csv(calculate_mean(read_data()))
```
